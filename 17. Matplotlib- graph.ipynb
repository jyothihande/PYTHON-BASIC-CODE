```python
x=[21,25,24,25,27,31,55]
y=[19,31,42,56,55,22,60]
plt.plot(x,y,marker='s',ms=14,mfc='b',mec='r',c='y',ls='--',lw=4)
plt.xlabel("age",c='r',loc='left')
plt.ylabel("salary",c='b',loc='bottom')
plt.title("Age and salary",loc='right')
plt.grid()
# shows gride behinde the graph use plt.grid(axis='y') for y
plt.show()
```

```python
example of scattered graph
```

```python
Class=[1,2,3,4,5,6,7,8,9]
Strength=[28,35,30,34,33,39,38,40,42]
plt.scatter(Class,Strength,marker='s')
plt.xlabel("Class",c='r')
plt.ylabel("Number of students",c='b')
plt.title("Number of students in class")
plt.grid(c='r')
plt.show()
```

```python
Class=[1,2,3,4,5,6,7,8,9]
class_strength_2012=[28,35,30,34,33,39,38,40,42]
class_strength_2022=[30,33,34,28,45,45,47,42,46]
class_strength_2032=[45,40,42,39,55,38,50,48,51]

plt.scatter(Class,class_strength_2012,marker='s')
plt.scatter(Class,class_strength_2022,c='y')
plt.scatter(Class,class_strength_2032,c='Purple',marker='*')

plt.xlabel("Class",c='r',fontsize=15)
plt.ylabel("Number of students",c='b',fontsize=15)
plt.title("Number of students in class",fontsize=15)
plt.grid(c='r')
plt.legend(['2012','2022','2023'],loc='upper right', bbox_to_anchor=(1, 1),ncol=3,facecolor='Pink',fontsize='10') # default was row
plt.show()
```
